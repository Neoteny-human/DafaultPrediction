```python
import os
import pandas as pd 
import numpy as np
```

```python
folder_path = './데이터'

file_list = os.listdir(folder_path)

data = pd.DataFrame()

for file in file_list:
    if file.endswith('.csv'):
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path, encoding='cp949')
        data = pd.concat([data, df], ignore_index=True)
```

```python
data
```

```python
data[data.duplicated(subset=['거래소코드', '회계년도'])]
```

```python
data[data.duplicated(subset=['거래소코드', '회계년도','감사의견코드'])]
```

```python
data = data.drop_duplicates(subset=['거래소코드', '회계년도'])
```

```python
def set_default_flag(row):
    if row['감사의견코드'] in ['QS', 'AG', 'DI', 'DS', 'DU']:
        return '1'
    else:
        return '0'
data['부실여부'] = data.apply(set_default_flag, axis=1)
data.dropna(inplace=True)
```

```python
#25만개의 정상
data[data['부실여부']=='0']
```

```python
#22000개의 부실
data[data['부실여부']=='1']
```

```python
data.to_csv('외감기업부실리스트.csv',encoding='utf-8-sig')
```
