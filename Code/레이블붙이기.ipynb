```python
import pandas as pd
df = pd.read_csv('./최종.csv',encoding='utf-8-sig',index_col=0)
df
```

```python
label = pd.read_csv('라벨최종.csv', encoding='utf-8-sig', index_col=0)
label
```

```python
df = df[df['회계년도'].str.endswith('/12')]
df
```

```python
df = df.merge(label[['회사명','회계년도','부실여부']], how='left', on=['회사명','회계년도'])
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df['부실여부'].value_counts()
```

```python
df['부실여부'] = df.groupby('회사명')['부실여부'].shift(1)
```

```python
df.dropna(inplace=True)
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df.index = pd.to_datetime(df['회계년도'])
df_revised = pd.DataFrame()
comp_list = list(df['회사명'].unique())

for i, comp in enumerate(comp_list):

    temp = df[df['회사명'] == comp]
    temp.sort_index(inplace=True)
    
    target = temp['부실여부']

    if ((target == 1).any()):
        default_index = list(target).index(1)
        tem = temp.iloc[:(default_index+1)]
    else:
        tem = temp

    if df_revised.empty != True:
        df_revised = pd.concat([df_revised, tem], axis=0)
    else:
        df_revised = tem
    
    print(f'{i}번째 읽어옴')
```

```python
df_revised['부실여부'].value_counts()
```

```python
df_revised.reset_index(drop=True,inplace=True)
# df_revised.to_csv('최종_동태.csv', encoding='utf-8-sig',index=False)
```

```python
df_revised.to_csv('최종+라벨링.csv',encoding='utf-8-sig',index=True)
```
