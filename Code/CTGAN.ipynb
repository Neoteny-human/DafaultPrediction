```python
from ctgan import CTGAN
import pandas as pd

# 데이터를 로드합니다.
data = pd.read_csv('./data/최종_동태_이상치처리후_train.csv')

# 학습할 때 조건을 걸 컬럼을 선택합니다.
conditional_columns = []

# CTGAN 모델을 초기화합니다.
ctgan = CTGAN()
data['부실여부']=data['부실여부'].astype(int)
data
```

```python
data = data.drop(['회사명','거래소코드','회계년도'],axis=1)
df_bad = data[data['부실여부'] == 1]
df_good = data[data['부실여부'] == 0]
```

```python
# 데이터를 학습합니다.
ctgan.fit(df_bad, conditional_columns)
```

```python
# 새로운 데이터를 생성합니다.
samples = ctgan.sample(10000)
samples
```

```python
minority_class = len(df_bad)
majority_class = len(df_good)
# 다수 클래스의 비율을 원하는대로 설정합니다.
desired_ratios = {1, 2, 4, 10}  # 10:1 비율로 설정 예시


for ratio in desired_ratios:
    # 오버샘플링할 샘플 수 계산
    num_samples_to_generate = int(majority_class / ratio - minority_class)
    
    # 샘플 생성
    sampled_minority_data = ctgan.sample(num_samples_to_generate)
    
    # 오버샘플링된 데이터와 majority 데이터 결합
    resampled_data = pd.concat([data, sampled_minority_data])
    
    # 결과 확인
    print(f"Class distribution after CTGAN Over Sampling with 1:{ratio} ratio:", resampled_data['부실여부'].value_counts())
    
    # 각 비율에 해당하는 데이터프레임을 CSV 파일로 저장
    resampled_data.to_csv(f'CTGAN_oversampled_train_1_{ratio}.csv', encoding='utf-8-sig' , index=False)
```
