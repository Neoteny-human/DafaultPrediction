```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('최종.csv',encoding='utf-8',index_col=0)
df.columns
```

```python
# df['부실여부'].value_counts()
df.to_csv('실험.csv',encoding='utf-8',index=False)
```

```python
# numeric_cols = df.select_dtypes(include=[np.number]).columns
numeric_cols = ['총자산경상이익률', '총자산순이익률', '금융비용/매출액', '금융비용/총부채',
       '순금융비용/매출액', '매출액경상이익률', '매출액순이익률', '자기자본경상이익률', '자기자본순이익률',
       'EBITDA/이자비용', 'EBIT/이자비용', '영업현금흐름/이자비용', '잉여현금흐름/총자산', '잉여현금흐름/총부채',
       '잉여현금흐름/이자비용', '부채상환계수', '자기자본비율', '유동자산/총자산', '이익잉여금/총자산', '이익잉여금/총부채',
       '이익잉여금/유동자산', '현금비율', '당좌비율', '유동비율', '유동부채회전율', '총자산', '매출액', '고정자산']

for col in numeric_cols:
    # 각 숫자형 컬럼에 대해 변화량 계산
    df[f'{col}_변화량'] = df.groupby('회사명')[col].diff()
    # NaN 값이 있는 행을 찾아 이전 2년간의 변화량을 계산하고, 2로 나눔
    mask = df[f'{col}_변화량'].isna()
    df.loc[mask, f'{col}_변화량'] = df.loc[mask].groupby('회사명')[col].diff(2) / 2

# NaN 값이 있는 행을 모두 삭제
df = df.dropna()
```

```python
df['부실여부'].value_counts()
df.to_csv('최종+동태.csv',index=False,encoding='utf-8-sig')
```
