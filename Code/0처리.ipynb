```python
import pandas as pd
import numpy as np

# 두 개의 예시 데이터프레임 생성
df1 = pd.DataFrame({
    'a': [1, 2, 3],
    'b': [4, 5, 6],
    'c': [7, 8, 9]
})

df2 = pd.DataFrame({
    'd': [10, 11, 12],
    'e': [13, 14, 15],
    'f': [16, 17, 18]
})

# 각 데이터프레임을 넘파이 배열로 변환
array1 = df1.to_numpy()
array2 = df2.to_numpy()

# 두 넘파이 배열을 세로로 이어붙이기
array_concat = np.vstack((array1, array2))

# 이어붙인 넘파이 배열을 다시 데이터프레임으로 변환하면서, df1의 열 이름을 사용
df_concat = pd.DataFrame(array_concat, columns=df1.columns)

df_concat
```
